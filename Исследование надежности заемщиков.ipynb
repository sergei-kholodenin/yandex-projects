## –®–∞–≥ 2. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤

```python
df.isna().sum()
```

```python
df.groupby('income_type').agg({'days_employed': ['count', 'min', 'max','median', 'mean'], 'total_income': ['min','max','median','mean'], 'debt': ['sum']})
```

–ü—Ä–æ–ø—É—Å–∫–∏ –µ—Å—Ç—å –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö days_employed –∏ total_income. –ù—É–∂–Ω–æ –∏—Ö –æ–±—Ä–∞–±–æ—Ç–∞—Ç—å. –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–∞–∂–∞ - —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ –æ—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–∏, —Ö–æ—Ç—è –∏–∑ —Å–≤–æ–¥–Ω–æ–π –≤–∏–¥–Ω–æ, —á—Ç–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ç–∞–∂ –≤—ã—à–µ —Ä–µ–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω–æ–≥–æ –∏ –≤—ã—Å—Ç–∞–≤–ª–µ–Ω —É –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–µ–Ω—Å–∏–æ–Ω–µ—Ä—ã –∏ –±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–µ. –ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–∏ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–∏ –≤–º–µ—Å—Ç–æ –¥–Ω–µ–π –≤ –¥–∞–Ω–Ω—ã–µ –±—ã–ª–∏ –≤–Ω–µ—Å–µ–Ω—ã —á–∞—Å—ã —Å—Ç–∞–∂–∞. –ù—É–∂–Ω–æ –æ—Ç–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞—Ç—å —ç—Ç–æ, —á—Ç–æ–±—ã –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–∞–Ω—á–µ–Ω–∏–µ –≤—ã—á–∏—Å–ª—è–ª–æ—Å—å –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ.

```python
def days_employed(row):
    income_type=row['income_type']
    days_employed=row['days_employed']
    if income_type =='–±–µ–∑—Ä–∞–±–æ—Ç–Ω—ã–π':
        return days_employed/24
    if income_type =='–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä':
        return days_employed/24
    return days_employed
df['days_employed']=df.apply(days_employed,axis=1)
df.head(15)
```

–ó–Ω–∞—á–µ–Ω–∏—è 0 –≤ –≤–æ–∑—Ä–∞—Å—Ç–µ - —è–≤–Ω–∞—è –æ—à–∏–±–∫–∞, –∑–∞–º–µ–Ω—è—é –Ω–∞ –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã–µ –ø—Ä–æ–ø—É—Å–∫–∞:

```python
dob_years_median = df['dob_years'].median()
df['dob_years']=df['dob_years'].replace(0, dob_years_median)
df.groupby('dob_years')['dob_years'].count()
```

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∫–æ—Ä—Ä–µ–∫—Ç–Ω–µ–π –≤—Å—Ç–∞–≤–∏—Ç—å –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ, —Ä–∞–∑–±–∏–≤–∞—é –Ω–∞ –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–µ –≥—Ä—É–ø–ø—ã —Å —à–∞–≥–æ–º 10 –ª–µ—Ç –¥–æ 65, –¥–∞–ª–µ–µ –ø–µ–Ω—Å–∏–æ–Ω–µ—Ä—ã.

```python
def years_group(years):
    if years <20:
        return '–¥–æ 20'
    if years <30:
        return '20-30'
    if years <40:
        return '30-40'
    if years <50:
        return '40-50'
    if years <60:
        return '50-60'
    if years <65:
        return '60-65'
    return '—Å—Ç–∞—Ä—à–µ 65'  
df['years_group']=df['dob_years'].apply(years_group)
df.head()
```

–ü–æ –∫–∞–∂–¥–æ–π –∏–∑ –≥—Ä—É–ø–ø —Å—á–∏—Ç–∞—é –º–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –∑–∞–º–µ–Ω—è—é –∏–º –ø—Ä–æ–ø—É—Å–∫–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ days_employed. –ü—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏–≤–æ–∂—É –ø–æ –º–æ–¥—É–ª—é.

```python
df['days_employed']=df['days_employed'].abs()
df['median_employed']=df.groupby('years_group')['days_employed'].transform('median')
df['days_employed']=df['days_employed'].fillna(df['median_employed'])
df.head(15)
```

–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –ø–æ –º–µ–¥–∏–∞–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –æ—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å —Ç–∞–±–ª–∏—Ü—É –∏ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –ø–æ –ø—Ä–æ–ø—É—Å–∫–∞–º.  –ü—Ä–æ–ø—É—Å–∫–∞ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö days_employed –∏ total_income —Å–æ–æ—Ç–≤–µ—Ç—Å–≤—É—é—Ç (2174 –∑–Ω–∞—á–µ–Ω–∏–π). –ò–∑ —á–µ–≥–æ –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –ø–æ –¥–∞–Ω–Ω—ã–º –∫–ª–∏–µ–Ω—Ç–∞–º –Ω–µ –±—ã–ª–æ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–æ —Å–≤–µ–¥–µ–Ω–∏–π –æ —Ä–∞–±–æ—Ç–µ, –ø–æ—ç—Ç–æ–º—É –Ω–µ –∏–∑–≤–µ—Å—Ç–µ–Ω —Å—Ç–∞–∂ –∏ –Ω–µ –∏–∑–≤–µ—Å—Ç–Ω–æ —Å—É–º–º—ã –¥–æ—Ö–æ–¥–∞.

```python
median_table=df[df['days_employed']== df['median_employed']]
median_table.groupby('income_type').agg({'days_employed': ['count', 'min', 'max','median', 'mean'], 'total_income': ['min','max','median','mean'], 'debt': ['sum']})
```

–ù—É–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–æ–∫ —Å—Ç–∞–∂–∞. –û—à–∏–±–∫–∞ –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∞–Ω–∞–ª–∏–∑–∞. –î–ª—è —ç—Ç–æ–≥–æ –≤–æ–∑—Ä–∞—Å—Ç –∑–∞ –≤—ã—á–µ—Ç–æ–º 18 –ª–µ—Ç (—Å–æ–≤–µ—Ä—à–µ–Ω–Ω–æ–ª–µ—Ç–∏–µ –∏ –Ω–∞—á–∞–ª–æ —Ç—Ä—É–¥–æ–≤–æ–π –¥–µ—è—Ç–µ–ª—å–Ω–æ—Å—Ç–∏) —è —É–º–Ω–æ–∂–∞—é –Ω–∞ 365 –¥–Ω–µ–π, –ø–æ—Å–ª–µ —á–µ–≥–æ –≤—ã—á–∏—Ç–∞—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π —Å—Ç–∞–∂–∞. –ü—Ä–∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–º –∑–∞–Ω—á–µ–Ω–∏–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ days_employed_error –æ—Ç—Ä–∞–∂–∞–µ—Ç—Å—è 1 (—Ç.–µ. —ç—Ç–æ –æ—à–∏–±–æ—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ).

```python
def days_employed_error(row):
    dob_years=row['dob_years']
    days_employed=row['days_employed']
    if (dob_years-18)*365-days_employed <0:
        return 1
    return 0
df['days_employed_error']=df.apply(days_employed_error, axis=1)
df.head(15)
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –∫–∞–∫–æ–≤–∞ –¥–æ–ª—è —Å—Ç—Ä–æ–∫ —Å –æ—à–∏–±–æ—á–Ω—ã–º —Å—Ç–∞–∂–µ–º –≤ days_employed:

```python
df['days_employed_error'].sum()/df['days_employed_error'].count()
```

```python
df['income_type']=pd.qcut(df['total_income'],
                          q=[0, 0.2, .4, .6, .8, 1],
                          labels=['–Ω–∏–∑–∫–∏–π','–Ω–∏–∂–µ —Å—Ä–µ–¥–Ω–µ–≥–æ','—Å—Ä–µ–¥–Ω–∏–π','–≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ','–≤—ã—Å–æ–∫–∏–π',])
df['income_type']=df['income_type'].fillna('–Ω–∏–∑–∫–∏–π')
```

–ü–æ –∞–Ω–∞–ª–æ–≥–∏–∏ —Å days_employed –ø—Ä–æ–≤–æ–∂—É –∑–∞–º–µ–Ω—É –ø—Ä–æ–ø—É—Å–∫–æ–≤ –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —É—Ä–æ–≤–Ω—è –¥–æ—Ö–æ–¥–æ–≤.

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

–¢–µ–ø–µ—Ä—å –Ω—É–∂–Ω–æ –≤ —Å—Ç–æ–ª–±—Ü–µ purpose –∑–∞–º–µ–Ω–∏—Ç—å —Ü–µ–ª–∏ –Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–µ. –î–ª—è —ç—Ç–æ–≥–æ –Ω—É–∂–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é.

```python
from nltk.stem import SnowballStemmer 
russian_stemmer = SnowballStemmer('russian')
russian_stemmer.stem('–∞–≤—Ç–æ–º–æ–±–∏–ª—å')

            
         
```

```python
def category_purpose(purpose):    
            if purpose.find('–∞–≤—Ç–æ–º–æ–±–∏–ª')>0:
                return '–∞–≤—Ç–æ–º–æ–±–∏–ª—å'
            if purpose.find('–æ–±—Ä–∞–∑–æ–≤–∞–Ω')>0:
                return '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'
            if purpose.find('—Å–≤–∞–¥—å–±')>0:
                return '—Å–≤–∞–¥—å–±–∞'
            return '–Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å'
df['category_purpose']=df['purpose'].apply(category_purpose)
df.head(15)
```

<div class="alert alert-success">
<h2>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</h2>

üëç
</div>

```python
category_purpose_group=df.groupby('category_purpose').agg({ 'debt': ['sum','count','mean']})
category_purpose_group.columns=['debt','total','%']
debt_total = category_purpose_group['debt'].sum() / category_purpose_group['total'].sum()
def risk_group(percent):
    if percent >= debt_total:
        return 'yes'
    return 'no'
category_purpose_group['risk_group'] = category_purpose_group['%'].apply(risk_group)
category_purpose_group
```
