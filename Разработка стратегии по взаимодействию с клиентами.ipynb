## –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤

–ü–æ—Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, –≥–¥–µ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ ‚Äî —Ñ–∞–∫—Ç –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –º–µ—Å—è—Ü–µ. –î–ª—è —ç—Ç–æ–≥–æ —Ä–∞–∑–æ–±—å–µ–º –≤—ã–±–æ—Ä–∫—É –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 80/20.

```python
X = df.drop(['churn'], axis=1)
y = df['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

<div class="alert alert-success">
<b>–£—Å–ø–µ—Ö:</b> –¢—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–ª –≤—ã–±–æ—Ä–∫–∏. –•–æ—Ä–æ—à–æ, —á—Ç–æ –Ω–µ –∑–∞–±—ã–ª –ø–∞—Ä–∞–º–µ—Ç—Ä random_state üëçüèª
–ú–æ–∂–Ω–æ —Ç–∞–∫–∂–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä stratify=y. –¢–æ–≥–¥–∞ –Ω–∞—à –¥–∞—Ç–∞—Å–µ—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç—Å—è  —Ç–∞–∫, —á—Ç–æ–±—ã –∏ –≤ —Ç—Ä–µ–π–Ω–µ –∏ –≤ —Ç–µ—Å—Ç–µ –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—à–µ–¥—à–∏—Ö\–æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∫–ª–∏–µ–Ω—Ç–æ–≤ –±—ã–ª–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π.</div>

<div class="alert alert-success">
<b>–£—Å–ø–µ—Ö:</b> –ö–æ–º–º–µ—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:

–î–æ–±–∞–≤–∏–ª, –ø–æ—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ —ç—Ç–æ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ –¥–∞–ª—å–Ω–µ–π—à–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.
</div>

–û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏:


*   –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π,
*   —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º.

```python
models = [
          LogisticRegression(random_state=0),
          RandomForestClassifier(n_estimators = 100, random_state=0)
]

```

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('accuracy:{:.2f} precision:{:.2f} recall:{:.2f}'.format(
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred)))
```

–û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏–º –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –ö–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫?

```python
for i in models:
  print(i)
  make_prediction(
        m=i,
        X_train=X_train,
        y_train=y_train,
        X_test=X_test,
        y_test=y_test
  )
```
