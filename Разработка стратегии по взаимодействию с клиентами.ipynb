# Разработка стратегии по взаимодействию с клиентами

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных.

Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Как понять, что клиент больше не с вами? Можно записать в отток тех, кто попросил закрыть договор или удалил аккаунт. Однако клиенты не всегда уходят демонстративно: чаще перестают пользоваться сервисом тихо.

Индикаторы оттока зависят от специфики отрасли. Когда пользователь редко, но стабильно закупается в интернет-магазине — не похоже, что он «отвалился». А вот если две недели не заходит на канал с ежедневно обновляемым контентом, дела плохи: подписчик заскучал и, кажется, оставил вас.

Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Конечно, не исключено, что он уехал на Бали и по приезде обязательно продолжит ходить на фитнес. Однако чаще бывает наоборот. Если клиент начал новую жизнь с понедельника, немного походил в спортзал, а потом пропал — скорее всего, он не вернётся.

Чтобы бороться с оттоком, отдел по работе с клиентами «Культуриста-датасаентиста» перевёл в электронный вид множество клиентских анкет. Моя задача — провести анализ и подготовить план действий по удержанию клиентов.
А именно:

- научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

     1) выделить целевые группы клиентов;
     
     2) предложить меры по снижению оттока;
     
     3) определить другие особенности взаимодействия с клиентами.

## Построение модели прогнозирования оттока клиентов

Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце. Для этого разобьем выборку на обучающую и валидационную в соотношении 80/20.

```python
X = df.drop(['churn'], axis=1)
y = df['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

<div class="alert alert-success">

Обучим модель на train-выборке двумя способами:


*   логистической регрессией,
*   случайным лесом.

```python
models = [
          LogisticRegression(random_state=0),
          RandomForestClassifier(n_estimators = 100, random_state=0)
]

```

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('accuracy:{:.2f} precision:{:.2f} recall:{:.2f}'.format(
        accuracy_score(y_test, y_pred),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred)))
```

Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке. Сравним по ним модели. Какая модель показала себя лучше на основании метрик?

```python
for i in models:
  print(i)
  make_prediction(
        m=i,
        X_train=X_train,
        y_train=y_train,
        X_test=X_test,
        y_test=y_test
  )
```

<div class="alert alert-success">
<b>Успех:</b> Отличная идея с функцией. Весь этап выполнен корректно.</div>

## Кластеризация клиентов

Стандартизируем данные, чтобы признаки были соизмеримы и веса пропорциональны.

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(X)
```

Построим матрицу расстояний функцией linkage() на стандартизованной матрице признаков и нарисуем дендрограмму.

```python
linked = linkage(X_sc, method = 'ward')

plt.figure(figsize=(15, 10))  
dendrogram(linked, orientation='top')
plt.title('Кластеризация клиентов')
plt.show() 
```

На основании по лученных данных можно сделать вывод, что клиентов можно разделить на 5 кластеров. Обучим модель кластеризации на основании алгоритма K-Means и спрогнозируем кластеры клиентов. Договоримся за число кластеров принять n=5. Посмотрим средние значения признаков по кластерам.

```python
km = KMeans(n_clusters=5, random_state=0)
labels = km.fit_predict(X_sc)
df['cluster_km'] = labels
print(df.groupby(['cluster_km']).mean())
```

Во втором кластере видно, что признак удаленности нулевой, т.е. объекты живут и работают далеко от места, практически нулевой показатель приглашения от друзей, самый низкий показатель посещения групповых занятий (21,5%). У второго и третьего кластера самый низкий показатель возраста, самый низкий показатель по продолжительности абонемента (около 2 месяцев), меньше всего платят за доп. услуги, самый низкий показатель по времени посещения в месяцах, частоте посещений в месяц и за последний месяц.
